```python
import numpy as np
```

```python
lattice = np.load("utilities/lattice.npy")
ldoms = np.load("utilities/l_doms.npy")
```

```python
indices = []
for i, (x, y, z) in enumerate(lattice):
    for j, (xd, yd, zd) in enumerate(ldoms):
        if (x, y, z) == (xd, yd, zd):
            indices.append((i,j))
```

```python
indices = np.asarray(indices)
```

```python
isin_lattice = np.isin(lattice, ldoms)
lattice_idx = np.argwhere(isin_lattice).flatten()
ldoms_idx = np.argsort(np.array([np.argwhere(ldoms == a) for a in lattice[isin_lattice]]).flatten())
doms_map_np = lattice_idx[ldoms_idx]
```

```python
doms_map = indices[np.argsort(indices[:, 1])][:, 0]
```

```python
doms_map
```

```python
doms_map_np
```

```python
np.all(doms_map == doms_map_np)
```
