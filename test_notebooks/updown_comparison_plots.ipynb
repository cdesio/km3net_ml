```python
updown_nocuts_true = np.loadtxt("utilities/updown_nocuts_true.txt")
updown_nocuts_pred = np.loadtxt("utilities/updown_nocuts_predicted.txt")
```

```python
updown_cuts_true = np.loadtxt("utilities/updown_cuts_true.txt")
updown_cuts_pred = np.loadtxt("utilities/updown_cuts_predicted.txt")
```

```python
def process_cosz(y):
    y_p = np.copy(y)
    y_p[y_p>0] = 1
    y_p[y_p<=0] = 0
    return y_p
```

```python
true_cosz_nocuts_lab = process_cosz(true_cosz_nocuts)
reco_cosz_nocuts_lab = process_cosz(reco_cosz_nocuts)
```

```python
true_cosz_cuts_lab = process_cosz(true_cosz_cuts)
reco_cosz_cuts_lab = process_cosz(reco_cosz_cuts)
```

```python
print(true_cosz_nocuts.shape, reco_cosz_nocuts.shape, pred_cosz_nocuts.shape)
```

```python
ROC_curve_tot(updown_nocuts_true, updown_nocuts_pred, to_categorical(updown_nocuts_pred),
             title ="ROC curve up/down Neural Network - no cuts")
```
