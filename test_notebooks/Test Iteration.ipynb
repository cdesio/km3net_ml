```python
from keras import backend as K
```

```python
def process_cosz(y):
    y[y>0] = 1
    y[y<=0] = 0
    return to_categorical(y)

def get_TZ_only(X):
    TZ = np.sum(X, axis=(2, 3))
    if K.image_data_format() == 'channels_first':
        TZ = TZ[:, np.newaxis, ...]
    else:
        TZ = TZ[..., np.newaxis]
    return TZ
```

```python
from get_iterations import get_n_iterations
```

### Generate TZ from one single file

```python
batch_size = 64
n_files = 1
iterations, tot_evts = get_n_iterations(fnames_train[:n_files], batch_size=batch_size)
data_gen = generator(fnames_train[:n_files], batch_size=batch_size, fdata=get_TZ_only, ftarget=process_cosz)
for i in range(iterations):
    try:
        X_batch, labels_batch = next(data_gen)
        print('='*40, '- {} - '.format(i+1), '='*40)
        print('Data Batch shape: ', len(X_batch), X_batch[0].shape)
        print('Labels Batch shape: ', labels_batch.shape)
        print('='*80)
    except StopIteration:
        pass
```

```python
def process_cosz(y):
    y[y>0] = 1
    y[y<=0] = 0
    return to_categorical(y)

def get_TZXY_only(X):
    TZ = np.sum(X, axis=(2, 3))
    XY = np.sum(X, axis=(1, 4))
    if K.image_data_format() == 'channels_first':
        TZ = TZ[:, np.newaxis, ...]
        XY = XY[:, np.newaxis, ...]
    else:
        TZ = TZ[..., np.newaxis]
        XY = XY[..., np.newaxis]
    return [TZ, XY]
```

```python
batch_size = 64
n_files = 2
iterations, tot_evts = get_n_iterations(fnames_train[:n_files], batch_size=batch_size)
data_gen = generator(fnames_train[:n_files], batch_size=batch_size, fdata=get_TZXY_only, ftarget=process_cosz)

for i in range(iterations):
    try:
        X_batch, labels_batch = next(data_gen)
        print('='*40, '- {} - '.format(i+1), '='*40)
        print('Data Batch shape: ', len(X_batch), X_batch[0].shape, X_batch[1].shape)
        print('Labels Batch shape: ', labels_batch.shape)
        print('='*80)
    except StopIteration:
        pass
```
