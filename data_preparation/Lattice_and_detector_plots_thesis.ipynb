```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from lattice_doms_znewk import lattice_doms
```

```python
lattice = np.load("utilities/lattice.npy")
ldoms = np.load("utilities/l_doms.npy")
```

```python
from detector_positions import structured_positions
```

```python
detector = "utilities/km3net_jul13_90m.detx"
```

```python
doms, pmts = structured_positions(detector)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(14,14))
ax = Axes3D(fig)
ax.scatter(doms["x"], doms["y"], doms["z"], alpha=0.35, s=450, 
           c="cornflowerblue", edgecolors="royalblue",
          label="DOMs positions")
ax.grid()
ax.set_xlabel('x(m)',size=15)
ax.set_ylabel('y(m)',size=15)
ax.set_zlabel('z(m)',size=15)
ax.tick_params(labelsize=13)
plt.legend(loc="upper right")
#ax.view_init(-1, 10)
#plt.show()
plt.savefig("plots/Detector_XYZ_view.png")
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter(doms["x"], doms["y"], s=200, 
            c="lightskyblue", edgecolors="cornflowerblue",
           label="DOMs pos")
plt.grid()
ax.set_xlabel('x(m)')
ax.set_ylabel('y(m)')
plt.title("Detector X-Y view")
plt.legend(loc="upper right")
#ax.view_init(30, 60)
#plt.show()
#plt.savefig("plots/Detector_XY_view.png")
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter(doms["x"], doms["y"], s=200, 
            c="lightskyblue", edgecolors="cornflowerblue",
           label="DOMs pos")
plt.plot(doms["x"], doms["y"], 
        label="DUs order")
plt.grid()
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.title("Detector X-Y view")
plt.legend(loc="lower right")
#plt.show()
plt.savefig("plots/Detector_XY_view_dus_ordered.png")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)
ax.scatter(lattice["x"], lattice["y"], lattice["z"], alpha=0.2, s=100)
ax.grid()
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('z (m)')
#ax.mouse_init(rotate_btn=1, zoom_btn=3)
#plt.show()
plt.savefig("plots/Lattice_XYZ_view.png")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)
ax.scatter(lattice["x"], lattice["y"], lattice["z"], alpha=0.30, s=50,label="Lattice pos")
ax.scatter(ldoms["x"], ldoms["y"], ldoms["z"], alpha=0.40, s=50, label="Lattice DOMs pos")
ax.grid()
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('z (m)')
plt.legend(loc="upper right")
#ax.view_init(30, 30)
#plt.show()
plt.savefig("plots/Lattice_and_detector_XYZ_view.png")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12,8))
plt.scatter(lattice["x"], lattice["y"], alpha=0.2, s=150, 
            c="aquamarine", edgecolors="seagreen",
           label="lattice pos")
plt.scatter(ldoms["x"], ldoms["y"], alpha=0.3, s=150, 
            c="steelblue", edgecolors="cornflowerblue",
           label="lattice DOMs pos")
plt.grid(linestyle="dotted")
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.title("Lattice and detector X-Y view")
#ax.view_init(30, 60)
#plt.show()
plt.legend(loc="upper right")
plt.savefig("plots/Lattice_and_Detector_XY_view.png")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
#plt.scatter(lattice["x"], lattice["y"], alpha=0.2, s=100)
plt.scatter(ldoms["x"], ldoms["y"], s=150, label="lattice doms", c="lightskyblue", edgecolors="cornflowerblue")
plt.scatter(doms["x"], doms["y"], s=150, label="real doms", c="salmon", edgecolors="indianred")
plt.grid(linestyle='dotted')
plt.xlabel('x (m)',size=14)
plt.xticks(np.arange(-540, 630, 90))
plt.ylabel('y (m)', size=14)
plt.yticks(np.arange(-540, 630, 90))
plt.legend(loc="upper right")
plt.title("DOMs vs 'Lattice DOMs' positions")
#ax.view_init(30, 60)
#plt.show()
plt.savefig("plots/DOMs_vs_Lattice_DOMs.png")
```

```python
plt.hist(np.abs(ldoms["x"]-doms["x"]), bins=20, color="salmon", edgecolor="indianred")#,histtype="step") 
plt.title("DOMs x-coordinate displacement")
plt.xlabel("(X_DOMs - X_Lattice_DOMs) (m)",size=12)
plt.show()
```

```python
plt.hist(np.abs(ldoms["y"]-doms["y"]), bins=20, color="salmon", edgecolor="indianred")# histtype="step", color="red") 
plt.title("DOMs y-coordinate displacement")
plt.xlabel("(Y_DOMs - Y_Lattice_DOMs)(m)",size=12)
plt.show()
```

```python
plt.hist(np.abs(ldoms["z"]-doms["z"]), bins=20, color="salmon", edgecolor="indianred")# histtype="step", color="red") 
plt.title("DOMs Z-coordinate displacement")
plt.xlabel("(Z_DOMs - Z_Lattice_DOMs)(m)",size=12)
plt.xlim(0,4)
plt.show()
```
